```python
import speech_recognition as sr
import pyttsx3
import math
import sys

engine = pyttsx3.init()

recognizer = sr.Recognizer()

def speak(text):
    engine.say(text)
    engine.runAndWait()

def listen():
    with sr.Microphone() as source:
        print("Listening for your command...")
        recognizer.adjust_for_ambient_noise(source)  
        audio = recognizer.listen(source)  
        print("Recognizing...")
        try:
            command = recognizer.recognize_google(audio)  
            print(f"You said: {command}")
            return command
        except sr.UnknownValueError:
            speak("Sorry, I did not understand that.")
            return None
        except sr.RequestError:
            speak("Sorry, the speech service is unavailable.")
            return None

def evaluate_expression(expression):
    try:
        result = eval(expression)
        return result
    except Exception as e:
        return f"Error: {str(e)}"

def calculator():
    speak("Welcome to Voice Activated Calculator. Say 'exit' to quit.")
    
    while True:
        speak("Please say your mathematical expression.")
        expression = listen()
        
        if expression:
            if "exit" in expression.lower():
                speak("Goodbye!")
                sys.exit()
            
            result = evaluate_expression(expression)
            if isinstance(result, (int, float)):
                result_text = f"The result is {result}"
                speak(result_text)
                print(result_text)
            else:
                speak(result)
                print(result)


if __name__ == "__main__":
    calculator()
```
